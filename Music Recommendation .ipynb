```python
import pandas as pd
import IPython.display as ipd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn import preprocessing
```

```python
feature30sec=pd.read_csv('./MCnR/features_30_sec.csv',index_col='filename')
feature30sec
```

```python
labels = feature30sec[['label']]
labels
```

```python
feature30sec=feature30sec.drop(columns=['length','label'])
feature30sec.head()
```

```python
scaledFeatures=preprocessing.scale(feature30sec)
print('Scaled Features type:', type(scaledFeatures), scaledFeatures.shape)
```

```python
similarFeature = cosine_similarity(scaledFeatures)
print("SimilarFeature shape:", similarFeature.shape)
```

```python
simFeatureLabels=pd.DataFrame(similarFeature)
simFeatureNames=simFeatureLabels.set_index(labels.index)
simFeatureNames.columns=labels.index
simFeatureNames.head()
```

```python
def findSimilarMusic(name):
    series = simFeatureNames[name].sort_values(ascending = False)
    series = series.drop(name)
    print("5 Similar Music to ", name)
    print(series.head(5))
```
