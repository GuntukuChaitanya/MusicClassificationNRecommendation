```python
import numpy as np
import librosa as lb
import librosa.display
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
audio, sampleRate = lb.load('blues0000.wav',sr=None)
plt.figure(figsize=(14,6))
librosa.display.waveplot(audio, color="#2B4F72")
ipd.Audio('blues0000.wav')
```

# Spectogram

```python
stft=lb.stft(audio)
stft_db=lb.amplitude_to_db(abs(stft))
```

```python
plt.figure(figsize=(14,6))
lb.display.specshow(stft,sr=sampleRate,x_axis='time',y_axis='hz')
plt.colorbar()
```

```python
plt.figure(figsize=(14,6))
lb.display.specshow(stft_db,sr=sampleRate,x_axis='time',y_axis='hz')
plt.colorbar()
```

```python
plt.figure(figsize=(14,6))
S=librosa.magphase(stft)[0]
lb.feature.rms(S=S)
lb.display.specshow(lb.amplitude_to_db(S, ref=np.max),
                         y_axis='log', x_axis='time')
plt.title('log Power spectrogram')
plt.show()
```

# Chromagram

```python
S = np.abs(stft)
chroma_stft = lb.feature.chroma_stft(S=S, sr=sampleRate)
chroma_stft
```

```python
plt.figure(figsize=(14,6))
plt.title('Chromagram')
lb.display.specshow(chroma_stft,x_axis='time',y_axis='chroma')
plt.colorbar()
```

# Root-Mean-Square

```python
S, phase = lb.magphase(stft)
rms = lb.feature.rms(S=S)
times = lb.times_like(rms)
plt.figure(figsize=(14,6))
plt.semilogy(times, rms[0], label='RMS Energy')
plt.xticks=[]
plt.legend()
```

# Zero Crossing Rate

```python
start=1000
end=1200
plt.figure(figsize=(14,6))
plt.plot(audio[start:end])
plt.grid()
```

```python
zeroCrossRate=lb.zero_crossings(audio[start:end],pad=False)
print("No.Of Zero Crossing: ",sum(zeroCrossRate))
```

# Roll-Off Frequency

```python
# Approximate maximum frequencies with roll_percent=0.99
rolloff = lb.feature.spectral_rolloff(y=audio, sr=sampleRate, roll_percent=0.99)
# Approximate minimum frequencies with roll_percent=0.01
rolloff_min = lb.feature.spectral_rolloff(y=audio, sr=sampleRate, roll_percent=0.01)
plt.figure(figsize=(14,6))
plt.plot(lb.times_like(rolloff),rolloff[0],label='Roll-off frequency (0.99)')
plt.plot(lb.times_like(rolloff),rolloff_min[0],color='black',
        label='Roll-off frequency (0.01)')
plt.legend()
plt.title('Roll-Off freqs')
```

# Harmonic and Percussive

```python
harmAudio, percAudio = lb.effects.hpss(audio)
plt.figure(figsize=(14,6))
librosa.display.waveplot(harmAudio, sr=sampleRate, alpha=0.25,label='Harmonic')
librosa.display.waveplot(percAudio, sr=sampleRate, color='r', alpha=0.5,label='Percussive')
plt.legend()
plt.title('Harmonic and Percussive')
```
